```python
from dataclasses import dataclass
```

```python
experiment_name = "EXP1.2.3"
task_tag = "exp1.2.3"


@dataclass
class S:
    start_x: float
    end_x: float


param_sets: list[S] = [
    S(start_x=50.0, end_x=50.0),  # gap 0 test
    # symmetrical
    S(start_x=1.5, end_x=98.5),
    S(start_x=2.5, end_x=97.5),
    S(start_x=3.0, end_x=97.0),
    S(start_x=5.0, end_x=95.0),
    S(start_x=10.0, end_x=90.0),
    S(start_x=20.0, end_x=80.0),
    S(start_x=30.0, end_x=70.0),
    S(start_x=40.0, end_x=60.0),
    S(start_x=45.0, end_x=55.0),
    S(start_x=49.0, end_x=51.0),
    # asymmetrical
    S(start_x=0.0, end_x=3.0),
    S(start_x=0.0, end_x=5.0),
    S(start_x=0.0, end_x=10.0),
    S(start_x=97.0, end_x=100.0),
    S(start_x=95.0, end_x=100.0),
    S(start_x=90.0, end_x=100.0),
]

param_sets_1: list[S] = [S(start_x=0.0, end_x=0.0), S(start_x=50.0, end_x=50.0), S(start_x=0.0, end_x=5.0)]

commands = [
    (
        f"./convert-notebook.sh notebooks/experiment-1.ipynb"
        f" --name='{experiment_name}-x({p.start_x}-{p.end_x})'"
        f" gap_start_x={p.start_x}"
        f" gap_end_x={p.end_x}"
        f" task_name='{experiment_name}: x=({p.start_x}-{p.end_x})'"
        f" task_tag='{task_tag}'"
    )
    for p in param_sets_1
]

print(" ; \\\n".join(commands))
```

# Good gamma calculation

We want to calculate some good values for the gamma. I observed that the learning rate at around 0.0003 to 0.0005 is the best one to start with, because higher ones lead to divergence from the beginning. Now, 

```python
experiment_name = "Stability 2"
task_tag = "stability-2"


@dataclass
class S:
    seed: int
    steps: int
    lr: float
    num_bins: int
    

steps = 10000
lrs = [0.0005, 0.0003]
num_bins = [10, 20]
seed = [420, 421, 422, 423, 424]

param_sets: list[S] = [
    S(seed=s, lr=l, steps=steps, num_bins=n)
    for l in lrs
    for n in num_bins
    for s in seed
]

commands = [
    (
        f"./convert-notebook.sh notebooks/experiment-binned.ipynb"
        f" --name='{experiment_name}-(seed={s.seed},bins={s.num_bins})'"
        f" task_name='{experiment_name}: seed={s.seed}, num_bins={s.num_bins}'"
        f" task_tag='{task_tag}'"
        f" seed={s.seed}"
        f" learning_rate={s.lr}"
        f" num_bins='{s.num_bins}'"
    )
    for s in param_sets
]

print(" ; \\\n".join(commands))
```
